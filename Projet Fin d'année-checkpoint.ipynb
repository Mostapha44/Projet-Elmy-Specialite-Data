```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données des pilotes
drivers = pd.read_csv("drivers.csv")

# Identifier les pilotes que nous voulons conserver
selected_drivers = ['Lewis Hamilton', 'Fernando Alonso', 'Sebastian Vettel', 'Max Verstappen', 'George Russell', 'Sergio Perez', 'Charles Leclerc', 'Carlos Sainz', 'Lando Norris', 'Oscar Piastri', 'Fernando Alonso', 'Lance Stroll', 'Pierre Gasly', 'Esteban Ocon', 'Daniel Ricciardo', 'Yuki Tsunoda', 'Valtteri Bottas', 'Guanyu Zhou', 'Kevin Magnussen', 'Nico Hülkenberg', 'Alexander Albon', 'Logan Sargeant']

# Créer une liste des noms complets des pilotes
driver_full_names = drivers['forename'] + ' ' + drivers['surname']

# Filtrer les données des pilotes pour ne garder que les pilotes sélectionnés
selected_drivers_data = drivers[driver_full_names.isin(selected_drivers)]

# Obtenir les identifiants des pilotes sélectionnés
selected_driver_ids = selected_drivers_data['driverId'].tolist()

# Fusionner les données des pilotes et des résultats des courses sur la colonne 'raceId'
merged_data = pd.merge(results, races, on='raceId')

# Grouper les résultats par année de course et par pilote, puis calculer les points obtenus par chaque pilote au cours de chaque saison
driver_season_points = merged_data.groupby(['year', 'driverId']).sum()['points'].reset_index()

# Sélectionner les saisons de 2014 à 2021
selected_seasons = driver_season_points[(driver_season_points['year'] >= 2008) & (driver_season_points['year'] <= 2016)]

# Sélectionner uniquement les pilotes sélectionnés
selected_seasons_selected_drivers = selected_seasons[selected_seasons['driverId'].isin(selected_driver_ids)]

# Créer un dictionnaire de correspondance entre les identifiants de pilote et leurs noms
driver_names = selected_drivers_data.set_index('driverId').loc[selected_driver_ids, ['forename', 'surname']]
driver_names['name'] = driver_names['forename'] + ' ' + driver_names['surname']

# Créer un graphique
plt.figure(figsize=(15, 7))

# Parcourir les pilotes sélectionnés et tracer leurs performances
for driver_id, name in zip(selected_driver_ids, driver_names['name']):
    driver_data = selected_seasons_selected_drivers[selected_seasons_selected_drivers['driverId'] == driver_id]
    plt.plot(driver_data['year'], driver_data['points'], marker='o', label=name)

# Ajouter des titres et des légendes
plt.title('Performance des pilotes sélectionnés de 2014 à 2021')
plt.xlabel('Année')
plt.ylabel('Points')
plt.legend(title='Pilote', bbox_to_anchor=(1.05, 1), loc='upper left')

# Afficher le graphique
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données des pilotes et des résultats des courses
drivers = pd.read_csv("drivers.csv")
results = pd.read_csv("results.csv")
races = pd.read_csv("races.csv")
circuits = pd.read_csv("circuits.csv")

# Identifier l'ID de Max Verstappen et Lewis Hamilton
max_verstappen_id = drivers[(drivers['forename'] == 'Max') & (drivers['surname'] == 'Verstappen')]['driverId'].iloc[0]
hamilton_id = drivers[(drivers['forename'] == 'Lewis') & (drivers['surname'] == 'Hamilton')]['driverId'].iloc[0]

# Filtrer les résultats des courses pour ne garder que les courses de la saison 2021 pour Verstappen et Hamilton
verstappen_results_2021 = results[(results['driverId'] == max_verstappen_id) & (results['raceId'].isin(races[races['year'] == 2021]['raceId']))]
hamilton_results_2021 = results[(results['driverId'] == hamilton_id) & (results['raceId'].isin(races[races['year'] == 2021]['raceId']))]

# Fusionner avec les données des courses pour obtenir les noms des circuits
verstappen_results_2021 = pd.merge(verstappen_results_2021, races[['raceId', 'circuitId', 'name']], on='raceId')
hamilton_results_2021 = pd.merge(hamilton_results_2021, races[['raceId', 'circuitId', 'name']], on='raceId')

# Créer un graphique
plt.figure(figsize=(20, 7))

# Tracer les résultats de Max Verstappen et Lewis Hamilton avec des lignes reliant les points
plt.plot(verstappen_results_2021['name'], verstappen_results_2021['positionOrder'], marker='o', label='Max Verstappen')
plt.plot(hamilton_results_2021['name'], hamilton_results_2021['positionOrder'], marker='o', label='Lewis Hamilton')

# Ajouter des titres et des légendes
plt.title('Résultats de course de Max Verstappen et Lewis Hamilton sur différents circuits en 2021')
plt.xlabel('Circuit')
plt.ylabel('Position de course')
plt.legend()

# Personnaliser l'axe y pour afficher les numéros de 1 à 20 de haut en bas sans .5
plt.yticks(range(1, 21))
plt.gca().invert_yaxis()

# Faire pivoter les noms des circuits pour une meilleure lisibilité
plt.xticks(rotation=45, ha='right')

# Afficher le graphique
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Données des tours et des arrêts au stand
tours = list(range(1, 68))  # Ajustement pour correspondre aux positions des pilotes
verstappen_positions = [1] * 67  # Max Verstappen part de la première position
hamilton_positions = [2] * 67  # Lewis Hamilton part de la deuxième position
bottas_positions = [3] * 67  # Valtteri Bottas part de la troisième position

# Insérer les tours des arrêts au stand
verstappen_stops = [24, 60]  # Tours où Verstappen a effectué ses arrêts
hamilton_stops = [28, 42]  # Tours où Hamilton a effectué ses arrêts
bottas_stops = [23, 53]  # Tours où Bottas a effectué ses arrêts

# Tracer le graphique des positions des pilotes
plt.figure(figsize=(12, 8))

# Trajectoire de Verstappen
plt.plot(tours, verstappen_positions, label='Verstappen', color='orange')

# Trajectoire de Hamilton
plt.plot(tours, hamilton_positions, label='Hamilton', color='green')

# Trajectoire de Bottas
plt.plot(tours, bottas_positions, label='Bottas', color='blue')

# Marquer les arrêts au stand
plt.scatter(verstappen_stops, [1] * len(verstappen_stops), color='orange', marker='x', label='Verstappen Pit Stops')
plt.scatter(hamilton_stops, [2] * len(hamilton_stops), color='green', marker='x', label='Hamilton Pit Stops')
plt.scatter(bottas_stops, [3] * len(bottas_stops), color='blue', marker='x', label='Bottas Pit Stops')

# Ajouter des titres et des légendes
plt.title('Positions des Pilotes lors du Grand Prix d\'Espagne 2021 avec Arrêts au stand')
plt.xlabel('Tours')
plt.ylabel('Positions')
plt.xticks(range(0, 68, 5))
plt.yticks([1, 2, 3])
plt.gca().invert_yaxis()  # Inverser l'axe y pour avoir la première position en haut
plt.legend(loc='upper right')

# Afficher le graphique
plt.grid(True)
plt.show()
```

```python
# Calculer l'évolution des positions gagnées pour chaque pilote
verstappen_position_gain = [0] * 67
hamilton_position_gain = [0] * 67
bottas_position_gain = [0] * 67

for i in range(1, 67):
    if i in verstappen_stops:
        verstappen_position_gain[i] = verstappen_position_gain[i-1] - 1
    else:
        verstappen_position_gain[i] = verstappen_position_gain[i-1]
    
    if i in hamilton_stops:
        hamilton_position_gain[i] = hamilton_position_gain[i-1] - 1
    else:
        hamilton_position_gain[i] = hamilton_position_gain[i-1]
    
    if i in bottas_stops:
        bottas_position_gain[i] = bottas_position_gain[i-1] - 1
    else:
        bottas_position_gain[i] = bottas_position_gain[i-1]

# Tracer le graphique de l'évolution des positions gagnées
plt.figure(figsize=(12, 8))

# Tracer l'évolution des positions gagnées pour chaque pilote
plt.plot(tours, verstappen_position_gain, linestyle='--', color='orange', label='Verstappen Position Gain')
plt.plot(tours, hamilton_position_gain, linestyle='--', color='green', label='Hamilton Position Gain')
plt.plot(tours, bottas_position_gain, linestyle='--', color='blue', label='Bottas Position Gain')

# Ajouter des titres et des légendes
plt.title('Évolution des Positions Gagnées lors du Grand Prix d\'Espagne 2021')
plt.xlabel('Tours')
plt.ylabel('Positions Gagnées')
plt.xticks(range(0, 68, 5))
plt.yticks(range(0, 3))
plt.legend(loc='upper right')

# Afficher le graphique
plt.grid(True)
plt.show()
```
