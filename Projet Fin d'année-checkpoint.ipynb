```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données des pilotes
drivers = pd.read_csv("drivers.csv")

# Identifier les pilotes que nous voulons conserver
selected_drivers = ['Lewis Hamilton', 'Fernando Alonso', 'Sebastian Vettel', 'Max Verstappen', 'George Russell', 'Sergio Perez', 'Charles Leclerc', 'Carlos Sainz', 'Lando Norris', 'Oscar Piastri', 'Fernando Alonso', 'Lance Stroll', 'Pierre Gasly', 'Esteban Ocon', 'Daniel Ricciardo', 'Yuki Tsunoda', 'Valtteri Bottas', 'Guanyu Zhou', 'Kevin Magnussen', 'Nico Hülkenberg', 'Alexander Albon', 'Logan Sargeant']

# Créer une liste des noms complets des pilotes
driver_full_names = drivers['forename'] + ' ' + drivers['surname']

# Filtrer les données des pilotes pour ne garder que les pilotes sélectionnés
selected_drivers_data = drivers[driver_full_names.isin(selected_drivers)]

# Obtenir les identifiants des pilotes sélectionnés
selected_driver_ids = selected_drivers_data['driverId'].tolist()

# Fusionner les données des pilotes et des résultats des courses sur la colonne 'raceId'
merged_data = pd.merge(results, races, on='raceId')

# Grouper les résultats par année de course et par pilote, puis calculer les points obtenus par chaque pilote au cours de chaque saison
driver_season_points = merged_data.groupby(['year', 'driverId']).sum()['points'].reset_index()

# Sélectionner les saisons de 2014 à 2021
selected_seasons = driver_season_points[(driver_season_points['year'] >= 2008) & (driver_season_points['year'] <= 2016)]

# Sélectionner uniquement les pilotes sélectionnés
selected_seasons_selected_drivers = selected_seasons[selected_seasons['driverId'].isin(selected_driver_ids)]

# Créer un dictionnaire de correspondance entre les identifiants de pilote et leurs noms
driver_names = selected_drivers_data.set_index('driverId').loc[selected_driver_ids, ['forename', 'surname']]
driver_names['name'] = driver_names['forename'] + ' ' + driver_names['surname']

# Créer un graphique
plt.figure(figsize=(15, 7))

# Parcourir les pilotes sélectionnés et tracer leurs performances
for driver_id, name in zip(selected_driver_ids, driver_names['name']):
    driver_data = selected_seasons_selected_drivers[selected_seasons_selected_drivers['driverId'] == driver_id]
    plt.plot(driver_data['year'], driver_data['points'], marker='o', label=name)

# Ajouter des titres et des légendes
plt.title('Performance des pilotes sélectionnés de 2014 à 2021')
plt.xlabel('Année')
plt.ylabel('Points')
plt.legend(title='Pilote', bbox_to_anchor=(1.05, 1), loc='upper left')

# Afficher le graphique
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

def get_race_data(season, round):
    url = f'http://ergast.com/api/f1/{season}/{round}/laps.json?limit=2000'
    response = requests.get(url)
    data = response.json()
    laps = data['MRData']['RaceTable']['Races'][0]['Laps']
    
    lap_data = {}
    
    for lap in laps:
        lap_number = int(lap['number'])
        for timing in lap['Timings']:
            driver = timing['driverId']
            position = int(timing['position'])
            
            if driver not in lap_data:
                lap_data[driver] = [None] * len(laps)
            
            lap_data[driver][lap_number - 1] = position
            
    df = pd.DataFrame(lap_data)
    df.index.name = 'Lap'
    return df

# Récupérer les données pour le Grand Prix d'Espagne 2021
season = 2021
round = 4
df = get_race_data(season, round)

# Afficher les colonnes pour identifier les noms exacts des pilotes
print(df.columns)

# Utiliser les noms exacts des pilotes
drivers_to_plot = ['hamilton', 'max_verstappen', 'bottas']  # Remplacez par les identifiants corrects

# Vérifier les colonnes disponibles et ajuster les noms si nécessaire
df_filtered = df[drivers_to_plot]

# Tracer le graphique
plt.figure(figsize=(18, 8))
for driver in df_filtered.columns:
    plt.plot(df_filtered.index + 1, df_filtered[driver], label=driver.capitalize())

plt.gca().invert_yaxis()
plt.xlabel('Tours')
plt.ylabel('Position')
plt.title('2021 Spanish Grand Prix - Position par Tour (Hamilton, Verstappen, Bottas)')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```
